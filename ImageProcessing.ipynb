```python
import matplotlib.pyplot as plt
from skimage import data,io
```

```python
img=data.chelsea()
io.imshow(img)
```

```python
from skimage.color import rgb2gray
grayimg=rgb2gray(img)
io.imshow(grayimg)
```

```python
from skimage.color import rgb2hsv
hsvimg=rgb2hsv(img)
io.imshow(hsvimg)
```

```python
from skimage.transform import resize
resizedimg=resize(img,(50,150))
io.imshow(resizedimg)
```

```python
from skimage.transform import rescale
rescaledimg=rescale(img,scale=(0.5,0.5))
io.imshow(rescaledimg)
```

```python
from skimage.transform import rotate
rotatedimg=rotate(img,angle=90)
io.imshow(rotatedimg)
```

```python
from numpy import fliplr
im_fliplr=fliplr(img)
io.imshow(im_fliplr)
```

```python
from numpy import flipud
im_flipud=flipud(img)
io.imshow(im_flipud)
```

```python
cropped=img[100:(img.shape[0]-50),100:(img.shape[1]-50)]
io.imshow(cropped)
```

```python
from skimage import exposure
img_bright=exposure.adjust_gamma(img,gamma=0.5,gain=1)
img_dark=exposure.adjust_gamma(img,gamma=2.5,gain=1)
plt.subplot(121),io.imshow(img_bright)
plt.subplot(122),io.imshow(img_dark)
```

```python
from skimage.filters import sobel_h
img_sobel=sobel_h(grayimg)
io.imshow(img_sobel)
```
